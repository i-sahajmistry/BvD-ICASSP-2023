```python
# Importing all necessary libraries
import cv2
import os
import glob
import pandas as pd
```

```python
MAIN_FOLDER = "./Dataset/Drone/train_videos"
TO_FOLDER   = "./Dataset/Drone/train_images"
```

```python
df = pd.read_csv(MAIN_FOLDER + '.txt', header=None)
df
```

```python
if not os.path.exists(TO_FOLDER):
    os.makedirs(TO_FOLDER)
```

```python
# Read the video from specified path
for video in df[0]:
    cam = cv2.VideoCapture(MAIN_FOLDER + '/' + video)
    try:
        IN_FOLDER = TO_FOLDER + '/' + video[:-4]
        # creating a folder named data
        if not os.path.exists(IN_FOLDER):
            os.makedirs(IN_FOLDER)

    # if not created then raise error
    except OSError:
        print ('Error: Creating directory of data')

    # frame
    currentframe = 0

    while(True):
        # reading from frame
        ret,frame = cam.read()

        if ret:
            # if video is still left continue creating images
            name = IN_FOLDER + '/frame_' + str(currentframe) + '.jpg'
            print ('Creating...' + name)

            # writing the extracted images
            cv2.imwrite(name, frame)

            # increasing counter so that it will
            # show how many frames are created
            currentframe += 5
        else:
            break

        for i in range(4):
            ret,frame = cam.read()

# Release all space and windows once done
cam.release()
cv2.destroyAllWindows()
```
